Scenario 5: Correlation Matrix for Housing Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Price': [200000, 250000, 180000, 300000, 400000, 350000, 280000],
    'Area_sqft': [1200, 1500, 1000, 1800, 2200, 2000, 1600],
    'Bedrooms': [2, 3, 2, 4, 5, 4, 3],
    'Bathrooms': [1, 2, 1, 3, 4, 3, 2],
    'Age_years': [10, 5, 15, 3, 2, 4, 7]
}

df = pd.DataFrame(data)
corr_matrix = df.corr()

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix - Housing Data", fontsize=14, fontweight='bold')
plt.show()
```

Scenario 6: Categorical Plot with Seaborn Themes

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

exam_data = {
    'Student': ['A','B','C','D','E','F','G','H','I','J','K','L'],
    'Subject': ['Math','Math','Science','Science','English','English',
                'Math','Science','English','Math','Science','English'],
    'Gender': ['Male','Female','Male','Female','Male','Female',
               'Female','Male','Male','Female','Male','Female'],
    'Score': [85, 90, 78, 88, 92, 80, 75, 89, 95, 82, 84, 87]
}

df_exam = pd.DataFrame(exam_data)

sns.set_theme(style="darkgrid", palette="pastel")
g = sns.catplot(data=df_exam, x="Subject", y="Score", hue="Gender", kind="box", height=5, aspect=1.2)
g.fig.suptitle("Exam Scores by Subject and Gender", fontsize=14, fontweight='bold')
g.set_axis_labels("Subject", "Score (out of 100)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
