```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/traffic_source.csv')

print(data.head())

plt.figure(figsize=(8,5))
bars = plt.bar(data['Source'], data['Visits'], color=['skyblue', 'orange', 'green', 'red', 'purple'])

plt.title('Website Traffic by Source')
plt.xlabel('Traffic Source')
plt.ylabel('Number of Visits')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

rain_data = pd.read_csv('/content/monthly_rainfall (2).csv')

print(rain_data.head())

colors = ['lightblue', 'lightblue', 'lightgreen', 'lightgreen', 'yellow', 'yellow',
          'orange', 'orange', 'red', 'red', 'lightgray', 'lightgray']

plt.figure(figsize=(10,5))
bars = plt.bar(rain_data['Month'], rain_data['Rainfall'], color=colors)

plt.title('Average Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(round(bar.get_height(),1)), ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
