**Visualize Relationships Between Variables Using Seaborn Pair**
**Plots**

**Scenario 6:** Wine Quality Assessment


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/winequality-red.csv", sep=';')

print("Dataset Preview:")
print(data.head())

features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']
wine_data = data[features]

sns.set(style="whitegrid", palette="Set2")
pair_plot = sns.pairplot(wine_data, hue='quality', diag_kind='kde', corner=True)

pair_plot.fig.suptitle("Pair Plot of Wine Chemical Properties vs Quality", y=1.02, fontsize=14)

plt.show()
```

**Scenario 5:**
 Student Performance Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Study_Hours': [2, 4, 3, 5, 6, 1, 7, 8, 2, 9, 3, 4, 6, 7, 5, 8, 9, 10, 4, 5],
    'Attendance_%': [60, 75, 68, 80, 85, 55, 90, 95, 65, 97, 70, 78, 88, 92, 81, 96, 98, 99, 73, 84],
    'Exam_Score': [50, 65, 58, 72, 80, 45, 85, 92, 55, 95, 60, 68, 79, 88, 74, 91, 94, 96, 67, 78],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female',
               'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

print("Dataset Preview:")
print(df.head())

print("\nCorrelation Matrix:")
print(df[['Study_Hours', 'Attendance_%', 'Exam_Score']].corr())

sns.set(style="whitegrid", palette="Set2")
pair_plot = sns.pairplot(df, hue='Gender', diag_kind='kde', corner=True)

pair_plot.fig.suptitle("Student Performance Analysis: Study Hours, Attendance, and Exam Scores", y=1.02, fontsize=14)
plt.show()
```
